```python
import xml.etree.ElementTree as ET
import os
from PIL import Image
import numpy as np

import matplotlib.pyplot as plt # for plotting

import torch # PyTorch package
import torchvision # load datasets
import torchvision.transforms as transforms # transform data
import torch.nn as nn # basic building block for neural neteorks
import torch.nn.functional as F # import convolution functions like Relu
import torch.optim as optim # optimzer
```

```python
def parse_xml_name(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()
    # Locate the <name> tag within the <object> tag
    traffic_light_name = root.find('.//object/name').text
    return traffic_light_name
```

```python
# get the y labels for our dataset
y = []
for i in range(1, 877, 1):
    y.append(parse_xml_name('annotations/road'+ str(i) +'.xml'))
```

```python
def pare_xml_boundary(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()
    
    bndbox = root.find('.//object/bndbox')
    xmin = int(bndbox.find('xmin').text)
    ymin = int(bndbox.find('ymin').text)
    xmax = int(bndbox.find('xmax').text)
    ymax = int(bndbox.find('ymax').text)
    return (xmin, ymin, xmax, ymax)
```

```python
pare_xml_boundary('annotations/road0.xml')
```

```python
def crop_image(file_path, bndbox):
    with Image.open(file_path) as img:
        cropped_image = img.crop(bndbox)
#         img_array = np.array(cropped_image)
        return cropped_image
```

```python
crop_image('images/road0.png', pare_xml_boundary('annotations/road0.xml'))
```

```python
crop_image('images/road1.png', pare_xml_boundary('annotations/road1.xml'))
```

```python
for i in range(1, 877, 1):
    crop_image('images/road'+ str(i) +'.png', pare_xml_boundary('annotations/road'+ str(i) +'.xml'))
```
